```python
import pandas as pd
```

# Get files

```python
# all phases
json_file = "/home/ahmed/Dropbox/DFCI/14_zoo/scrap-science/0_combined/data_180221.json"
json_df = pd.read_json(json_file)
print json_df.shape
json_df.head()
```

```python
# csv from gdoc
gdoc_file = "/home/ahmed/Dropbox/DFCI/14_zoo/rescience/gdoc/180312.csv"
gdoc_df = pd.read_csv(gdoc_file)
print gdoc_df.shape
gdoc_df.head()
```

# Filter

```python
# only phase == 1
json_df = json_df[json_df['phase']==1]
print json_df.shape
```

```python
# only use == 1
gdoc_df = gdoc_df[gdoc_df['use']==1.0]
print gdoc_df.shape
```

```python
# cleanup id column
gdoc_df['key'] = [x.split('.')[0].encode() for x in gdoc_df['id']]
```

# Combine by id

```python
combine = pd.merge(json_df, gdoc_df, how="inner", on="key")
print combine.shape
combine.head()
```

```python
# save
combine.to_csv("combined_180311.csv", encoding="utf-8")
```
